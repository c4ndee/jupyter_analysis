```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
epa_cars_raw = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-15/big_epa_cars.csv",
                          sep=',', error_bad_lines=False, index_col=False, dtype='unicode')

epa_cars = epa_cars_raw[['cylinders', 'city08','highway08']]
```

```python
epa_cars.describe()
```

```python
epa_cars.dtypes
```

```python
epa_cars["city08"] = pd.to_numeric(epa_cars["city08"])
epa_cars["highway08"] = pd.to_numeric(epa_cars["highway08"])
```

```python
plt.figure(figsize=(10,8))

fig = sns.scatterplot(data=epa_cars, x="city08", y="highway08", hue='cylinders', 
                palette=sns.color_palette("Set1", 9))

fig.set(xlabel="City MPG", ylabel="Highway MPG")
```

```python
plt.figure(figsize=(10,8))

g = sns.FacetGrid(epa_cars, col='cylinders', hue='cylinders', col_wrap=3)
g = g.map(sns.distplot, 'city08', bins=30, kde=False)
```
